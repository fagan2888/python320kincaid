```python
from PIL import Image
from matplotlib import pyplot as plt
from scipy import stats
import cv2
import numpy as np
```

```python
# first make a list of all the pixels
img = Image.open('blob1.jpg')
plt.imshow(img)
plt.show()
```

```python
width, height = img.size
p = 0
newPixel = [0] * (width*height)
for pixel in img.getdata():
    newPixel[p] = (pixel[0] * .8) + (pixel[1] * .8) + (pixel[2] * .1) # flatten RGB into one digit: arbitrarily chose these numbers to make the "orange" stand out more
    p += 1
```

```python
newImg = np.reshape(newPixel,[width,height])
plt.imshow(newImg)
plt.show()
np.unique(newImg)
```

```python
orangeArea = 0
otherArea = 0

for pixel in newPixel:
    if pixel > 270 and pixel < 290:
        orangeArea += 1
    else:
        otherArea += 1
# print('other area=' + str(otherArea)+', orangeArea='+str(orangeArea))
percentOrange = (orangeArea / otherArea) * 100
print('Image is ' + str(round(percentOrange,2)) +'% orange')
```
