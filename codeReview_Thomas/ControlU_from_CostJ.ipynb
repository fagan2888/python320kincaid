```python
import numpy as np
from scipy.integrate import odeint
import scipy.interpolate as interp  
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import time 
```

```python
def matrix_fillnan(Mat):
    isnn = np.isnan(Mat)  
    z_max =  Mat[~isnn].max()  
    z_max_log = np.log(z_max)  
    
    Mat_clean = Mat.copy()  # is there a way to avoid this? 
    Mat_clean[isnn] = z_max
    
    Mat_log = np.log(Mat_clean + 1 )   
    return Mat_clean, Mat_log
 
def pendulum_ode(y_n,t, u_n= 0    ): 
    g = -10. 
    L = 1. 
    mp = 1.
    fy = np.zeros_like(y_n)  
    time_sign =  1
    fy[0]= time_sign*  y_n[1]   
    fy[1] = time_sign* (  g/L*np.sin( y_n[0] )  + 1/(mp*L**2)*u_n )
    return fy

def calc_costToGo(x,u_j,Q=1,R=1, g = -10, E_d = 10): 
    E = 0.5*x[1]**2 + np.cos(x[0])*g 
    E_er = E-Ed 
    J_to_go = E_er*Q*E_er + u_j*R*u_j  # quadratic cost  
    return J_to_go 

def wrap2periodic(theta, period = 2*np.pi, center = 0 ):
    offset = np.pi - center
    thetaWrapped = (theta + offset ) %period - offset 
    return thetaWrapped 
  
```

```python
# ----------load file -----------
file_to_load = 'J_interp_Q1_R1.npy'
J_raw = np.load( file_to_load ) 

#  -----------create mesh -----------
y_lim = 15
n_y, n_x = np.shape( J_raw)  
x = np.linspace(-np.pi,np.pi, n_x   ) 
y = np.linspace(-y_lim,y_lim, n_y,endpoint = True ) 
X,Y = np.meshgrid(x,y)

# ----------- fill nan's and take log  -----------
J_clean, J_log = matrix_fillnan(J_raw) 
 
# ---------controller ------------
R = 1
Q = 1
u_clip = 6 
u_opts = np.array([-u_clip ,0, u_clip  ]) 
Ed = 10
#--------- time  ------------
dt = 0.05;  
t = np.linspace(0,dt,2)    

#  -----------plot  data -----------
f,  axs   = plt.subplots(1,1 ,figsize=(8,8) )    
axs .pcolor(X,Y , J_log , cmap='BuGn' )   

fig = plt.figure()
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X , Y ,  J_clean,  
                 cmap=cm.coolwarm , edgecolor='none')
ax.view_init(elev=50, azim=10)
ax.set_title('Cost J')
 
fig = plt.figure()
ax = fig.gca(projection='3d')  
surf = ax.plot_surface(X, Y, J_log   , cmap=cm.coolwarm , edgecolor='none')  
ax.view_init(elev=50, azim=10)
ax.set_title('Log of cost J  ')
```

```python
# ----------Interpolate 2d data with spline   
spline = interp.interp2d( x,y,J_clean.transpose().ravel() ,kind = 'cubic' )  # works 
ZI = spline( x,y)  
   
# ----------plot original 
fig = plt.figure()
ax = fig.gca(projection='3d') 
surf = ax.plot_surface(X , Y ,  J_clean,  
                 cmap=cm.coolwarm , edgecolor='none')
ax.view_init(elev=50, azim=10)
ax.set_title('Cost J')

# ----------plot interpolated data to check if interpolation is accurate---------------------
fig = plt.figure()
ax = fig.gca(projection='3d')  
surf = ax.plot_surface(X, Y, ZI  , cmap=cm.coolwarm , edgecolor='none')  
ax.view_init(elev=50, azim=10)
ax.set_title('Interpolated cost J  ') 
```

```python
#------------------------------------------------------
# run simulation on every grid point, 
# to see which of 3 control actions is cheapest
#------------------------------------------------------ 
 
U = np.full( np.shape(J_clean), np.nan  )
t_start = time.time()  
            
for j in range(len(x)):
    for k in range(len(y)): 
        yi = np.array( [X[k,j] , Y[k,j]] ) 
        Ji = J_clean[k,j ]
           
        cost_new_place = np.zeros_like(u_opts)  
        for jj, u_c in enumerate(u_opts, start=0):  
            #----------------solve simulation 
            y_n  = odeint(  pendulum_ode, yi,  t, args=(u_c,)  )[1]    # ode solver 
            y_n[0] = wrap2periodic(y_n[0],2*np.pi, 0 )                  # wrap theta to range of -pi to pi 

            # find cost of the step
            J_n  = calc_costToGo(yi,u_c, Q,R )  * dt

            # compute cost at new step
            cost_new_place[jj] = spline( y_n[0],y_n[1]  )   + J_n

        # compare costs and find which control action is cheapest
        ind_low = np.argmin(cost_new_place)  
        U[k,j] = u_opts[ind_low]  
             
print('runtime: ' + str( round(time.time()-t_start,3)) + ' seconds')  
```

```python
# plot results   
xh = np.linspace(- np.pi,np.pi,101)
yh = np.sqrt( 20*(1+np.cos(xh))  ) 
 
f, axs = plt.subplots(1,2 ,figsize=(16,8) )   

axs[0].pcolor(X,Y, J_log , cmap='BuGn' )  
axs[1].pcolor(X,Y, U  )   
   
axs[0].plot(xh,yh,'y' ,color = 'k')
axs[0].plot(xh,-yh,'y',color = 'k') 
axs[0].scatter( 0,0,color= 'yellow') 
axs[1].plot(xh,yh,'y' ,color = 'k')
axs[1].plot(xh,-yh,'y',color = 'k') 
axs[1].scatter( 0,0,color= 'yellow')  
```
