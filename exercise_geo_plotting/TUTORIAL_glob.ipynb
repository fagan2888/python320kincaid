```python
# PLOTS EACH DAY DATA FILE AND SAVE AS A SEPARATE FILE

for day in day_data:
    savename = "./figures/glob/"+day.split("/")[-1].split("\\")[-1].split(".csv")[0]+"_figure.png"
    title = day.split("/")[-1].split("\\")[-1].split(".csv")[0].replace("_"," ")
    
    df = pd.read_csv(day)
    
    fig, ax = plt.subplots()
    ax.set_title(title)
    ax.plot(df["x"], df["y"], marker='o')
    plt.savefig(savename)
```
