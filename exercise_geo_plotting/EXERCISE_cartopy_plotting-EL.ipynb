```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cartopy.io.shapereader as shpreader

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./data/final_loc_data.csv")
df.time = pd.to_datetime(df.time)
df.head()
```

```python
df.describe()
```

```python
temp = df[df.time == pd.to_datetime('1990-01-01')]

fig = plt.figure(figsize=(10, 7))
ax = plt.axes(projection=ccrs.Mercator())
ax.scatter(360-temp['lon'], temp['lat'], transform=ccrs.PlateCarree(), 
           c=temp['temp'], s=temp['precip'], cmap='magma', 
           vmin=df["temp"].min(), vmax=df["temp"].max())
# ax.add_feature(cfeature.COASTLINE)
ax.gridlines()

sm = plt.cm.ScalarMappable(cmap="magma",
        norm=plt.Normalize(df["temp"].min(), df["temp"].max()))
sm._A = []
plt.colorbar(sm,ax=ax)
plt.show()
```
