```python
import glob
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
files = sorted(glob.glob('./StackAsImageSequence/*.tif'))
```

```python
stack = []
for i, file in enumerate(files): 
    temp = cv2.imread(file)
    temp = cv2.cvtColor(temp, cv2.COLOR_BGR2GRAY)
    stack.append(temp)

stack = np.asarray(stack)
print(stack.shape[0], 'stacked copies of image that is', stack.shape[1:], 'pixels')

# Max values across stack
plt.figure()
plt.imshow(np.max(stack, axis=(0)))

# Mean values across stack
plt.figure()
plt.imshow(np.mean(stack, axis=(0)))

# Median values across stack
plt.figure()
plt.imshow(np.median(stack, axis=(0)))
```
